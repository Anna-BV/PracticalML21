#  Формулы динамической системы
*Аттрактор Лоренца*

\begin{align}
\dot{x} & = \sigma(y-x) \\
\dot{y} & = \rho x - y - xz \\
\dot{z} & = -\beta z + xy
\end{align}

```python
import numpy as np
from numpy import linspace, array
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
#значения параметров
s=10.0
r=28.0 
b=8.0/3.0
#Аттрактор Лоренца
def lorenz(t,xyz):
    x, y, z = xyz
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return [x_dot, y_dot, z_dot]
#начальное условие
y0 = [0, 1, 0]

solution = solve_ivp(lorenz, [0, 40], y0, t_eval = linspace(0, 40, 5000))
fig = plt.figure()

AL = fig.gca(projection = '3d')

AL.plot(solution.y[0], solution.y[1], solution.y[2], 'pink')
AL.set_xlabel("X ось")
AL.set_ylabel("Y ось")
AL.set_zlabel("Z ось")
AL.set_title("Аттрактор Лоренца")

plt.show()

```
